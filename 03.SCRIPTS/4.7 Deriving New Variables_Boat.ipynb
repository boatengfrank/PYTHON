```python
#define the function 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
#applying the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#counting the values 
df['price_range'].value_counts(dropna = False)
```

```python
#checking for the most expensive product
df['prices'].max()
```

## if statements with loc function

```python
#using if statmeents with the loc function 
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## new column being created to return busiest day, least busy day and regularly busy day 

```python
#using loop to create a column for the days and how busy they are 
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#combining result with the data frame
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna =False) 
```

## task 4. 7   #2. 

```python
#using loop to create new columns for busiest days and slowest days
results = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        results.append("Busiest days")
    elif value == 1:
        results.append("Busiest days")
    elif value == 3:
        results.append("Slowest days")
    elif value == 4:
        results.append("Slowest days")
    else:
        results.append("Regularly busy")
```

```python
#combining result with the data frame
df_ords_prods_merged['busiest_day'] = results
```

```python
#checking the frequency
df_ords_prods_merged['busiest_day'].value_counts(dropna =False)
```

## we can see from the frequency that regulary busy has the highest frequncy followed by busiest days being saturday and sunday

# identifying busiest hour of the day

```python
#finding frequency of the order hour of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create a for-loop if statement 
hour_of_day = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Add result to dataframe as "busiest_period_of_day" column
df_ords_prods_merged['busiest_period_of_day'] = hour_of_day
```

```python
# Checking if work is accurate 
df_ords_prods_merged[['order_hour_of_day', 'busiest_period_of_day']].head(10)
```
