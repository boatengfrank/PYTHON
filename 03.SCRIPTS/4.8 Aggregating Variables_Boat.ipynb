# # Creating loyalty flag for the existing customers

```python
# Group data by "user_id" column, generate maximum orders for each user, 
#create "max_order column" for aggregation results
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the data grouping for accuracy
pd.options.display.max_rows = None
```

```python
df_ords_prods_merged.head(100)
```

```python
# Creating flags to assign "loyalty" labels to user ID based on max order value
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 
                     'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking frequencies of loyalty flags for accuracy
df_ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

# Finding the mean of product prices by loyalty groups

```python
# Group by "prices" 
df_ords_prods_merged.groupby('prices')
```

```python
# Spliting the data into groups by "loyalty_flag" and then calculate the mean for each "prices" column
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean']})
```

# Creating spending flag for each user based on average price of products across orders

```python
# Group data by "user_id" column, generate mean product price for each user,
#create "mean_price" for aggregation results
df_ords_prods_merged['mean_price'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check grouping
df_ords_prods_merged.head(100)
```

```python
# Creating flags to assign "spending" labels to user ID based on mean product price
df_ords_prods_merged.loc[df_ords_prods_merged['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequencies of spending flags for accuracy
df_ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

# Creating order frequency flag that indicates users' ordering behavior

```python
# Group data by "user_id" column, generate median product of "days since prior order, 
#create "median_days" for aggregation results
df_ords_prods_merged['median_days'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check grouping
df_ords_prods_merged.head(100)
```

```python
# Creating flags to assign "freqency" labels to user ID based on median days since prior order
df_ords_prods_merged.loc[df_ords_prods_merged['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['median_days'] > 10) & (df_ords_prods_merged['median_days'] <= 20),
                     'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequencies of spending flags for accuracy
df_ords_prods_merged['frequency_flag'].value_counts(dropna=False)
```

```python
# Examining the null values
print(df_ords_prods_merged[df_ords_prods_merged.frequency_flag.isnull()])
```
